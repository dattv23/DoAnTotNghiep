```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
reviews_df = pd.read_csv('../02.Dataset/reviews_300k.csv')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="rating", data=reviews_df, palette="viridis")
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
user_review_counts = reviews_df["userId"].value_counts()
plt.figure(figsize=(6, 4))
sns.histplot(user_review_counts, bins=range(1, 30), kde=False)
plt.title("Number of Reviews per User")
plt.xlabel("Number of Reviews")
plt.ylabel("Number of Users")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
product_review_counts = reviews_df["productId"].value_counts()
plt.figure(figsize=(6, 4))
sns.histplot(
    product_review_counts, bins=range(1, 10), kde=False
)
plt.title("Number of Reviews per Product")
plt.xlabel("Number of Reviews")
plt.ylabel("Number of Products")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
